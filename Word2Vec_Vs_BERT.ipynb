```python
# from torch.nn.modules import distance
import numpy as np
from flair.embeddings import WordEmbeddings, TransformerWordEmbeddings
from flair.data import Sentence
from scipy.spatial import distance
```

#Word2Vec - Glove Embedding

```python
glove_embedding = WordEmbeddings('glove')
#we know that word2vec model creates vector for a word that is independent of the context.
#i.e, Numeric representation of a word will remain same irrespective of the context of a particular word.
#Glove is also a WordEmbedding model
```

##Sentence 1

```python
sentence_1 = Sentence("The founder of Apple is Steve Jobs")
```

```python
glove_embedding.embed(sentence_1)
```

```python
for token in sentence_1:
  print(token)
  print(token.embedding)
  print("\n")
```

##Glove distance between the same word

```python
#We know that in sent_1 --> Apple is pointing towards an organization.
#And in sent_2 --> It is pointing towards an apple.
```

```python
print(sentence_1)
print(sentence_2)
```

```python
glove_dist = distance.euclidean(np.array(sentence_1[3].embedding),
                                np.array(sentence_2[1].embedding))
print("Distance between Apple embeddings for Glove = {}".format(glove_dist))
```

```python
#Takeaway:

#So word2vec model don't understand the context of the word. To prove the hypothesis, we have taken the vectors represnetation of Apple in both the sentences.
#Lets calculate the eculidean distance of the word Apple in two sentences.
#Zero difference. So word2vec is context independent,i.e. never understand the context.
```
