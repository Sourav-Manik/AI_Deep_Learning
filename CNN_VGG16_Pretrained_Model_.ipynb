Import libraries and dataset

```python
from tensorflow import keras
from keras.utils import load_img
```

```python
model = keras.applications.VGG16(include_top = True,
                                weights = "imagenet",
                                input_tensor = None,
                                input_shape=None,
                                pooling = None,
                                classes = 1000,
                                classifier_activation = "softmax")
```

```python
model.summary()
```

```python
image = load_img(path = '/content/german shepard.jpeg', target_size = (224,224,3))
```

```python
image
```

```python
import numpy as np
np.array(image)
```

```python
image = np.reshape(a= image,newshape = (1,224,224,3))
image.shape
```

```python
from keras.applications.imagenet_utils import preprocess_input
```

```python
image_preprocessed = preprocess_input(image)
image_preprocessed
```

```python
y_pred = model.predict(image_preprocessed)
y_pred
```

```python
from keras.applications.vgg16 import decode_predictions
```

```python
decode_predictions(y_pred)
```
