```python
import numpy as np
import matplotlib.pyplot as plt
from transformers import pipeline
from pylab import rcParams
```

```python
rcParams['figure.figsize'] = 10,5
```

```python
zero_shot_classifier = pipeline("zero-shot-classification") #To download model
```

Zero shot classification is the process of using these pretrained models and architectures to perform classification for a set of labels that the model hasn’t been fine tuned to look for

```python
result = zero_shot_classifier(sequences = "I have to meet Mr. Arun in few minutes",
                              candidate_labels = ['urgent','not urgent','movies','food'],
                              multi_label = True)
result
```

```python
plt.bar(result["labels"],result["scores"])
plt.yticks(list(np.arange(0,1,0.1)))
plt.show()
```

```python
result = zero_shot_classifier(sequences = "I have to meet Mr. Arun in few minutes. But I can sapre 5 minutes for you. Let's talk.",
                              candidate_labels = ['urgent','not urgent','movies','food'],
                              multi_label = True)
result
```

```python
plt.bar(result["labels"],result["scores"])
plt.yticks(list(np.arange(0,1,0.1)))
plt.show()
```
